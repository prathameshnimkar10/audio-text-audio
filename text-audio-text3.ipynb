```python
from transformers import pipeline
import gradio as gr

music_genre_classifier = pipeline("audio-classification", model="SeyedAli/Musical-genres-Classification-Hubert-V1", device=0)

def classify_music_genre(audio_file):
    # Perform genre classification
    result = music_genre_classifier(audio_file)
    predictions = [{"Genre": res["label"], "Confidence": round(res["score"], 3)} for res in result]
    return predictions[0]

interface = gr.Interface(
    fn=classify_music_genre,
    inputs=gr.Audio(type="filepath", label="Upload Music File"),
    outputs=gr.JSON(label="Genre Predictions with Confidence Scores")
    )

# Launch the Gradio interface
interface.launch(debug = True)
```
