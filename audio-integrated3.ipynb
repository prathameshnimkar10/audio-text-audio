```python
import whisper
from transformers import Wav2Vec2FeatureExtractor, Wav2Vec2ForSequenceClassification, pipeline
from gtts import gTTS
import gradio as gr
import torch
import torchaudio
import os
from pydub import AudioSegment

# Load models
device = 0 if torch.cuda.is_available() else -1
asr_model = whisper.load_model("base")
feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("superb/wav2vec2-base-superb-er")
emotion_model = Wav2Vec2ForSequenceClassification.from_pretrained("superb/wav2vec2-base-superb-er").to(device)
conversation_model = pipeline("text2text-generation", model="google/flan-t5-large", device=device)

def convert_to_wav(file_path):
    try:
        if not file_path.endswith(".wav"):
            audio = AudioSegment.from_file(file_path)
            wav_path = "converted_temp.wav"
            audio.export(wav_path, format="wav")
            return wav_path
        return file_path
    except Exception as e:
        print(f"Error converting file to WAV: {e}")
        return None

def load_and_resample_audio(audio_file):
    try:
        waveform, sample_rate = torchaudio.load(audio_file)
        if sample_rate != 16000:
            resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000).to(device)
            waveform = resampler(waveform.to(device))
        return waveform.cpu().numpy(), 16000
    except Exception as e:
        print(f"Error loading or resampling audio: {e}")
        return None, None

def get_combined_transcription(audio_file, chunk_length=50):
    file_path = convert_to_wav(audio_file)
    if not file_path:
        return "Error: Unable to process audio file."

    audio = AudioSegment.from_wav(file_path)
    chunk_length_ms = chunk_length * 1000
    chunks = [audio[i:i + chunk_length_ms] for i in range(0, len(audio), chunk_length_ms)]

    combined_transcription = ""
    for i, chunk in enumerate(chunks):
        temp_path = "temp_chunk.wav"
        chunk.export(temp_path, format="wav")
        result = asr_model.transcribe(temp_path)
        combined_transcription += result['text'] + " "
        os.remove(temp_path)  # Clear temporary file after each chunk

    return combined_transcription.strip()

def analyze_audio_tone(audio_file):
    audio_input, _ = load_and_resample_audio(audio_file)
    if audio_input is None:
        return "Error: Unable to analyze tone."

    emotion_model = Wav2Vec2ForSequenceClassification.from_pretrained("superb/wav2vec2-base-superb-er").to(device)
    feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("superb/wav2vec2-base-superb-er")
    inputs = feature_extractor(audio_input, sampling_rate=16000, return_tensors="pt", padding=True).to(device)

    with torch.no_grad():
        audio_logits = emotion_model(**inputs).logits

    predicted_id = torch.argmax(audio_logits, dim=-1).cpu().numpy()[0]
    predicted_emotion = emotion_model.config.id2label[predicted_id]

    emotion_model.to("cpu")
    torch.cuda.empty_cache()
    return predicted_emotion

def generate_response(user_prompt, transcription, audio_emotion=None):
    input_text = (
        f"Transcription: {transcription}\n"
        f"Emotion: {audio_emotion if audio_emotion else 'unknown'}\n"
        f"User question: {user_prompt}\n"
        "Generate a relevant response:"
    )
    response = conversation_model(input_text, max_length=600, do_sample=True)[0]['generated_text']
    return response

def text_to_audio(text):
    tts = gTTS(text=text, lang='en')
    tts.save("response.mp3")
    return "response.mp3"

def process_audio_and_text(input_audio=None, input_text=None):
    if not input_audio and not input_text:
        return "No input provided", {"Error": "No audio or text input provided"}, None

    transcription = get_combined_transcription(input_audio) if input_audio else input_text
    if not transcription:
        return "No transcription available", {"Error": "No transcription could be generated"}, None

    audio_emotion = analyze_audio_tone(input_audio) if input_audio else "unknown tone"
    response_text = generate_response(input_text, transcription, audio_emotion)
    audio_response = text_to_audio(response_text)
    return transcription, {"Generated Response": response_text}, audio_response

# Gradio Interface
with gr.Blocks(theme='Ocean') as demo:
    with gr.Row():
        audio_input = gr.Audio(type="filepath", label="Upload Audio File")
        optional_text = gr.Textbox(lines=2, label="Optional Text Input (e.g., 'What is the mood?' or 'Give me insights')")

    with gr.Row():
        audio_transcription = gr.Textbox(label="Transcription", placeholder="Audio transcription will appear here.")
        output_insights = gr.JSON(label="Generated Response")
        output_audio = gr.Audio(label="Audio Response")

    audio_submit = gr.Button("Process Audio and Text")
    audio_submit.click(fn=process_audio_and_text, inputs=[audio_input, optional_text], outputs=[audio_transcription, output_insights, output_audio])

demo.launch(debug=True)
```
